# Split Colored Channel
 

```python
import numpy as np
```

```python
img1 = cv2.imread("/content/drive/MyDrive/IMP301/PT1/images/bao.jpg")
```

```python
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
#plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
# split the Blue, Green and Red color channels
blue,green,red = cv2.split(img1)

# define channel having all zeros
zeros = np.zeros(blue.shape, np.uint8)

# merge zeros to make BGR image
blueBGR = cv2.merge([blue,zeros,zeros])
greenBGR = cv2.merge([zeros,green,zeros])
redBGR = cv2.merge([zeros,zeros,red])
```

```python
# display the three Blue, Green, and Red channels as BGR image
plt.imshow(cv2.cvtColor(blueBGR,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(greenBGR,cv2.COLOR_BGR2RGB))
```

```python

plt.imshow(cv2.cvtColor(redBGR,cv2.COLOR_BGR2RGB))
```
